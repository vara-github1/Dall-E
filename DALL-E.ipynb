# **Generating image**



In order to generate an image using the DALLÂ·E we can use the openai.Image.create call.

````python
response = openai.Image.create(
  prompt= "a chef shark making shrimp.",
  n=1,
  size="256x256"
)
````

````python
import shutil

def copy_image(source_path, destination_path):
    try:
        shutil.copy2(source_path, destination_path)
        print("Image copied successfully!")
    except IOError as e:
        print(f"An error occurred while copying the image: {e}")

# Example usage
source_image = "original.png"
destination_image = "original-copy.png"
copy_image(source_image, destination_image)
````

The purpose of a mask is to define the region of an image that you want to change or apply an effect to, while leaving the rest of the image unchanged.

writing over our image


````python
## Load the original image
img = Image.open('original-copy.png')
# Convert the image to RGBA format
img = img.convert('RGBA')
````

````python
# Create a new blank image for the mask
mask = img.copy()
````

Now for the fun part we simply gonna draw a rectangle box over where the edit should be.

# **Chaining**


Chaining API calls provides us with the ability to produce a series of images in a sequential order using a sequence of prompts.

it's just like comics and manga you read

````python
# Set the prompts
prompts = ["dog eating food in from his bowl", "dog laying on bed upside down", "dog cuddling his human"]

# Generate and save the images
for i, prompt in enumerate(prompts):
    response = openai.Image.create(
      prompt=prompt,
      n=1,
      size="256x256"
    )

    # Get the image URL from the response
    image_url = response['data'][0]['url']

    # Download and save the image
    img_data = requests.get(image_url).content
    with open(f"a_dog_day{i+1}.jpg", 'wb') as handler:
        handler.write(img_data)
````

#**Synchronous Vs Asynchronous**

In a regular, synchronous program, tasks need to finish one by one before the next one can start. This can be wasteful, especially with tasks involving waiting for things like network responses, where the program just sits idle.
Async programming lets you do other things while waiting for tasks like network operations to complete, making better use of your time and resources.
Python's asyncio library makes it easy to write code that can do many things at once using coroutines. It's like juggling multiple tasks, such as handling network connections, without the need for separate threads or processes.

To measure the execution time of your code, you can use the `time.time()` function to capture the start time before the code execution and the end time after it finishes. ***Asynch Example***

Inorder to make Asynchronous API calls we will use Pythonâs asyncio and aiohttp libraries.

1. The `fetch_and_save_image` function is used to download and save an image from a given URL. It takes a session (aiohttp's ClientSession), a URL, and a file path as parameters. It uses aiohttp's `get()` to make a GET request to the URL and asynchronously reads the response content. Then, it opens a file at the specified path and writes the image data to it.


```
async def fetch_and_save_image(session, url, path):
    try:
        async with session.get(url) as resp:
            img_data = await resp.read()
            with open(path, 'wb') as handler:
                handler.write(img_data)
    except Exception as e:
        print(f"Unexpected error: {e}")
```



2. The `fetch_image` function sends a request to OpenAI's API to create an image based on the provided prompt. It requires a prompt and an index as input. The function utilizes `openai.Image.create()` to generate an image and retrieves the image URL from the response. It subsequently establishes an aiohttp ClientSession and invokes `fetch_and_save_image()` to download and store the image.


```
async def fetch_image(prompt, i):
    try:
        response = openai.Image.create(
            prompt=prompt,
            n=1,
            size="256x256"
        )

        # Get the image URL from the response
        image_url = response['data'][0]['url']

        # Download and save the image
        async with aiohttp.ClientSession() as session:
            await fetch_and_save_image(session, image_url, f"robot_dog_journey_{i+1}.jpg")
    except openai.api_errors.APIError as e:
        # This will catch any error returned by the OpenAI API
        print(f"API error: {e}")

    except Exception as e:
        # This is a catch-all for any other exceptions
        print(f"Unexpected error: {e}")
```



3. The `main` function leverages asyncio to create tasks for each prompt, adds them to a list, and concurrently executes them using asyncio.gather(). This approach initiates all tasks simultaneously, rather than one after another, which improves efficiency. It also measures execution time by recording the start and end times, enabling you to calculate how long the tasks took to run.



```
async def main():
    start_time = time.time()  # Start measuring execution time

    tasks = []
    for i, prompt in enumerate(prompts):
        tasks.append(fetch_image(prompt, i))

    await asyncio.gather(*tasks)

    end_time = time.time()  # End measuring execution time
    execution_time = end_time - start_time
    print(f"Execution time: {execution_time} seconds")
```



Please note that the openai.Image.create function is a blocking operation and doesnât support async. The code above assumes that this function is asynchronous, which is not correct. To truly parallelize this, youâd need an async-compatible OpenAI library or use something like Pythonâs concurrent.futures to run the blocking parts in a separate thread.

To remove the error i used that if not - else block in the end

````python
import os
import openai
import aiohttp
import asyncio
import secret
import time

# Set the prompts
prompts = ["cat eating food from his bowl", "cat laying on bed upside down", "cat cuddling his human"]

async def fetch_and_save_image(session, url, path):
    try:
        async with session.get(url) as resp:
            img_data = await resp.read()
            with open(path, 'wb') as handler:
                handler.write(img_data)
    except Exception as e:
        print(f"Unexpected error: {e}")

async def fetch_image(prompt, i):
    try:
        response = openai.Image.create(
            prompt=prompt,
            n=1,
            size="256x256"
        )

        # Get the image URL from the response
        image_url = response['data'][0]['url']

        # Download and save the image
        async with aiohttp.ClientSession() as session:
            await fetch_and_save_image(session, image_url, f"a_cat_day{i+1}.jpg")
    except openai.api_errors.APIError as e:
        # This will catch any error returned by the OpenAI API
        print(f"API error: {e}")
    except Exception as e:
        # This is a catch-all for any other exceptions
        print(f"Unexpected error: {e}")

async def main():
    start_time = time.time()  # Start measuring execution time

    tasks = []
    for i, prompt in enumerate(prompts):
        tasks.append(fetch_image(prompt, i))

    await asyncio.gather(*tasks)

    end_time = time.time()  # End measuring execution time
    execution_time = end_time - start_time
    print(f"Execution time: {execution_time} seconds")

if __name__ == "__main__":
    loop = asyncio.get_event_loop()
    if not loop.is_running():
        loop.run_until_complete(main())
    else:
        print("An event loop is already running.")

````
